```python
import pandas as pd
import numpy as np
import matplotlib as plt
import csv
from csv import writer #allows us to add rows in csv file
from io import StringIO
from pandas import *
from array import *
import disney_package # import package
%matplotlib inline

#lets first makes sure there no empty cells in our dataset 

df = pd.read_csv("/content/disney_movies.csv")
df[['mpaa_rating']] = df[['mpaa_rating']].fillna("Not Rated") #select Column mpa_rating replace all NaN with "Not Rated"
df[['genre']] = df[['genre']].fillna("No Genre") #select Column genre replace all NaN with "No Genre"
df.to_csv('noNullDisney.csv', index=False) #lets save this new dataset to a new name " noNullDisney"
df.drop
df.head()
```

```python
#since we have certain movies that either don't have a rating or genre or both we have 4 different sets of data 
#lets first extract the data set that contain only movies that have values for all columns
df = df[df.mpaa_rating != "Not Rated"] # removes all rows that contain "Not Rated" in rating column
df = df[df.genre != "No Genre"] #removes all rows that contain "No Genre" in genre column 
df.to_csv('noEmptyDisney.csv', index=False) #Save dataset as new csv noEmptyDisney
df.head()
```

```python
filename = open('/content/disney_movies.csv', 'r')
file = csv.DictReader(filename)
movieTitle = []
releaseDate = []
genre =[]
rating = []
totalGross = []
adjustedInflat = []

for col in file:
  movieTitle.append(col['movie_title'])
  releaseDate.append(col['release_date'])
  genre.append(col['genre'])
  rating.append(col['mpaa_rating'])
  totalGross.append(col['total_gross'])
  adjustedInflat.append(col['inflation_adjusted_gross'])

print(movieTitle)
print(releaseDate)
print(genre)
print(rating)
print(totalGross)
print(adjustedInflat)
```

```python
disneyMovies = np.vstack((movieTitle, releaseDate, genre, rating, totalGross, adjustedInflat)) #using vstack() from numpy we stack long the columns
np.set_printoptions(threshold=10000, linewidth=1000)
print(disneyMovies) 

```

```python
a = [row[342] for row in disneyMovies] #captures index row from 2D array using for loop, number indicate row number
b = [row[235] for row in disneyMovies]
c = [row[321] for row in disneyMovies]
d = [row[179] for row in disneyMovies]
e = [row[275] for row in disneyMovies]
f = [row[342] for row in disneyMovies]
a1 = " , ".join(a) #this allows us to convert the list into a string seperated by commas 
b1 = " , ".join(b)
c1 = " , ".join(c)
d1 = " , ".join(d)
e1 = " , ".join(e)
f1 = " , ".join(f)
dis = open("data.txt", "w") 
dis.writelines(a1)
dis.write("\n") #adds next line 
dis.writelines(b1)
dis.write("\n")
dis.writelines(c1)
dis.write("\n")
dis.writelines(d1)
dis.write("\n")
dis.writelines(e1)
dis.write("\n")
dis.writelines(f1)
dis.close()
```
