```python
#Storing the columns in a and b
a = df['total_gross']
b = df['inflation_adjusted_gross']
```

```python
#Plotting Main plot and sub plot
fig = plt.figure(figsize=(8,7))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(a,b)
axes1.set_title("Main Plot")
axes1.set_xlabel("total gross")
axes1.set_ylabel("inflation adjusted gross")

axes2.plot(a,b)
axes2.set_title("Sub Plot")
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))
axes[0].plot(a,b)
axes[0].set_title('First Plot')

axes[1].plot(b,a)
axes[1].set_title('Second Plot')
plt.tight_layout()
fig.savefig('my_plot.png',dpi=200)
```

```python
#Plotting a graph with some variations
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,a*2, label ='value*2',color='purple', linewidth=2,alpha=0.5,linestyle='--', marker='D',
          markersize=10,markerfacecolor ='yellow', markeredgewidth=3,markeredgecolor='blue')
axes.plot(a,a*10, label='value*10',color='blue', lw=2,alpha=0.5,ls='--', marker='o' )
axes.set_xlabel('Gross amount*2')
axes.set_ylabel('Gross amount*10')
axes.set_title("Different operations on gross amount")
axes.legend(loc = 0)
```

```python
#Here we are extracting the year from release_date and storing it into a new column
df['release_year'] = pd.DatetimeIndex(df['release_date']).year

#Then we are calculating the mean of the adjusted gross per genre and per year
group = df.groupby(['genre','release_year']).mean()

#Converting the groupby object to a dataframe
genre_yearly = group.reset_index()

#Visualizing genre_yearly dataframe
genre_yearly.head(10)
```

```python
import seaborn as sns
# Plot the above data  
sns.relplot(kind='line', x='release_year', y='inflation_adjusted_gross', hue='genre', data=genre_yearly)
```
